### PREPROCESSING DATA

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("IPL Matches 2008-2020.csv")
print(dataset.shape)
```

```python
#no need for umpire columns
dataset = dataset.iloc[:,0:-2]
```

```python
dataset
```

```python
dataset.sort_values(['id']).head(60)
```

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
for features in features_with_na : 
    print(features,np.round(dataset[features].isnull().mean(),4), '% missing value')
```

```python
#venues for which city is null
venues_city = pd.unique(dataset[dataset['city'].isnull()]['venue'])
print('Venues for which City value is null : ',venues_city)
```

```python
#checking for each venue ,what are the city values given other then null
unique_city = {}
for ven in venues_city:
    unique_city[ven] = pd.unique(dataset[(dataset['venue'] == ven)&(dataset['city'].notnull())]['city'])
    print("City for ",ven," : ",unique_city[ven])
```

```python
#from above we can say there is a unique value of city 
#So we are updating null values with the unique value of city
for ven in venues_city : 
    dataset.loc[dataset['venue'] == ven,'city'] = unique_city[ven][0]
```

```python
#missing value because due to rain match halted or canceled (winner, result, player_of_match,eliminator)
dataset[dataset['winner'].isnull()]
```

```python
#replacing each attribute null value with ''
dataset.loc[dataset['winner'].isnull(),'winner'] = ''
dataset.loc[dataset['result'].isnull(),'result'] = ''
dataset.loc[dataset['player_of_match'].isnull(),'player_of_match'] = ''
dataset.loc[dataset['eliminator'].isnull(),'eliminator'] = ''
```

```python
#result_margin null value due to rain, tie
dataset[dataset['result_margin'].isnull()].head()
```

```python
#replacing null value with -1
dataset.loc[dataset['result_margin'].isnull(),'result_margin'] = -1
```

```python
#method null for non duckworth-loius
dataset[dataset['method'].isnull()].head()
```

```python
#add column in TEAMS relation lastMatchOn
```

```python
TEAMS = pd.DataFrame()
TEAMS.name = 'TEAMS'
TEAMS['name'] = info.index
TEAMS['createdON'] = info.values
TEAMS['lastMatchOn'] = dataset.groupby('team1').last()['date'].apply(lambda x : x.split('-')[0]).values
TEAMS.reset_index(inplace=True)
TEAMS.rename({'index':'teamID'},axis=1,inplace=True)
```

```python
#homeTeamID attribute of LOCATION will be of format 'teamID1-teamID2-..' move from old home team to recent home team
LOCATION = pd.DataFrame()
LOCATION.name = 'LOCATION'
LOCATION['venue'] = info.index
LOCATION['city'] = info.values
LOCATION.reset_index(inplace=True)
LOCATION.rename({'index':'venueID'},axis=1,inplace=True)
home_team_idlst = []
for ven in LOCATION['venue'] : 
    tmp = home_team[ven].split('-')
    t_id = '-1'
    for t in tmp : 
        if(t != '') : 
            if(t_id != '-1') : t_id = t_id + "-" + str(TEAMS[TEAMS['name'] == t]['teamID'].values[0])
            else : t_id = str(TEAMS[TEAMS['name'] == t]['teamID'].values[0])
    home_team_idlst.append(t_id)
LOCATION['homeTeamID'] = home_team_idlst
```

```python
MATCHES = pd.DataFrame()
MATCHES.nam = 'MATCHES'
MATCHES['matchID'] = dataset['id']
MATCHES['date'] = dataset['date']
MATCHES['venueID'] = dataset['venue'].apply(lambda x : LOCATION[LOCATION['venue'] == x]['venueID'].values[0])
MATCHES['team1'] = dataset['team1'].apply(lambda x : TEAMS[TEAMS['name'] == x]['teamID'].values[0])
MATCHES['team2'] = dataset['team2'].apply(lambda x : TEAMS[TEAMS['name'] == x]['teamID'].values[0])
MATCHES['tossWin'] = dataset['toss_winner'].apply(lambda x : TEAMS[TEAMS['name'] == x]['teamID'].values[0])
MATCHES['tossDecision'] = dataset['toss_decision']
MATCHES['playerOfMatch'] = dataset['player_of_match']
MATCHES['winner'] = dataset['winner'].apply(lambda x : x if(x == '') else TEAMS[TEAMS['name'] == x]['teamID'].values[0])
MATCHES['result'] = dataset['result']
MATCHES['resultMargin'] = dataset['result_margin']
```

```python
BALL_BY_BALL = pd.DataFrame()
BALL_BY_BALL.name = 'BALL BY BALL'
BALL_BY_BALL['matchID'] = dataset1['id'].values
BALL_BY_BALL['inning'] = dataset1['inning'].values
BALL_BY_BALL['overs'] = dataset1['over'].values
BALL_BY_BALL['ball'] = dataset1['ball'].values
BALL_BY_BALL['batsman'] = dataset1['batsman'].values
BALL_BY_BALL['nonStriker'] = dataset1['non_striker'].values
BALL_BY_BALL['bowler'] = dataset1['bowler'].values
BALL_BY_BALL['batsmanRuns'] = dataset1['batsman_runs'].values
BALL_BY_BALL['extraRuns'] = dataset1['extra_runs'].values
BALL_BY_BALL['totalRuns'] = dataset1['total_runs'].values
BALL_BY_BALL['isWicket'] = dataset1['is_wicket'].apply(lambda x : True if(x == 1) else False).values
BALL_BY_BALL.reset_index(inplace=True)
BALL_BY_BALL.rename({'index':'id'},axis=1,inplace=True)
```
