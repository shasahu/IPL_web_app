5.1. For each team you need to give the stadium name(venue) where that team have won the maximum match?

```python
query = '''select t.name,l.venue,count(*) as win
           from teams as t,location as l,matches as m 
           where (m.venueID=l.venueID) and (m.winner is not NULL) and (m.winner=t.teamID) 
           group by t.name,l.venue'''
result = pd.read_sql(query,sql_engine)
```

```python
show = result[result['name'] == 'Chennai Super Kings'][['venue','win']]
```

```python
idx = result.groupby(['name'])['win'].transform(max) == result['win']
```

```python
from random import randint
```

```python
#random colors list created for each bar plot 
my_color = []
n = result[idx].shape[0]
for i in range(n):
    my_color.append('#%06X' % randint(0, 0xFFFFFF))
```

```python
plt.figure(figsize = (45,10))
for i in range(n) : 
    plt.bar(result[idx].iloc[i,0],result[idx].iloc[i,2],label=result[idx].iloc[i,1],color=my_color[i])
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (45,10))
plt.pie(result[idx].iloc[:,0],result[idx].iloc[:,2],labels=result[idx].iloc[:,1],color=my_color[i])
plt.legend()
plt.show()
```
